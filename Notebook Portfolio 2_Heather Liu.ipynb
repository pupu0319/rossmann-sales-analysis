# 3. Variable Explanation

The following are explanation about variable that utilized in the following analysis.

* Id - an Id that represents a (Store, Date) duple within the test set
* Store - a unique Id for each store
* Sales - the turnover for any given day (this is what you are predicting)
* Customers - the number of customers on a given day
* Open - an indicator for whether the store was open: 0 = closed, 1 = open
* StateHoliday - indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
* SchoolHoliday - indicates if the (Store, Date) was affected by the closure of public schools
* StoreType - differentiates between 4 different store models: a, b, c, d
* Assortment - describes an assortment level: a = basic, b = extra, c = extended
* CompetitionDistance - distance in meters to the nearest competitor store
* CompetitionOpenSince - gives the approximate year and month of the time the nearest competitor was opened
* Promo - indicates whether a store is running a promo on that day
* Promo2 - Promo2 is a continuing and consecutive promotion for some stores: 0 = store is not participating, 1 = store is participating
* Promo2Since - describes the year and calendar week when the store started participating in Promo2
* PromoInterval - describes the consecutive intervals Promo2 is started, naming the months the promotion is started anew. E.g. "Feb,May,Aug,Nov" means each round starts in February, May, August, November of any given year for that store

2. The chosen criteria can be following considerations:  
  
Mean: Filling the missing values with the mean of the 'competition distance' can be a simple and straightforward approach. It is suitable when the data is approximately normally distributed and there are no significant outliers.  

Median: Using the median can be more robust to outliers compared to the mean. If the 'competition distance' variable is skewed or contains outliers, using the median may be a better option.

3. So I decide to plot the CompetitionDistance distribution graph to see whether there're a number of obvious outliers. If so, I will use Median to replace missing values. Otherwise, I'll use Mean instead.

## 9.2 Question 2

**On the line graph, plot all the sales for store 1, store 25 and store 827 that occurred in December of 2014.**

```python
# Step 1: filtering out the DataFrame of Dec. 2014
dec = merged_sales_promo.loc[(merged_sales_promo['Dates'] > '2014-11-30')&(merged_sales_promo['Dates'] < '2015-1-1')]
dec.head()
```

```python
# Step 2: Filtering out the selected three stores' Sales figures
selected_store = dec.loc[(dec['Store']==1)|(dec['Store']==25)|(dec['Store']==827)]
selected_store
```

```python
sales = selected_store[['Store','Sales','Dates','Day']]
sales
```

```python
# Step 4: pivot rows 'Store' data into columns
import pandas as pd
pivot_sales = sales.pivot(index='Dates',columns=['Store'],values='Sales')
pivot_sales
```

```python
# Step 3: Plotting into line graph
pivot_sales.plot(kind='line',title='Sales From Three Stores in 2014 Dec', figsize=(15,5))
plt.xlabel('December 2014')
plt.ylabel('Sales')
plt.show()
```

Based on the graph, it is evident that three stores exhibit a relatively similar pattern of sales performance. Firstly, all three stores share the same trading hours, remaining closed on Sundays (7th, 14th, 21st, and 28th), as well as on Christmas Holidays (25th and 26th). Notably, Store number 827 stands out as the best performer among the three, achieving the highest sales records on Saturdays throughout the initial three weeks of December 2014, just before Christmas. Following a two-day holiday break (25th and 26th), sales experienced another peak on the 27th of December, and this trend continued from the 29th of December onwards. On the other hand, Store number 1 demonstrated the least favorable performance among the trio, generating approximately half of the sales compared to Store number 827.
